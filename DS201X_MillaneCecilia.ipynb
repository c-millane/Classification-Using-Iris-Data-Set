```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
import matplotlib.pyplot
%matplotlib inline 
```

```python
iris = load_iris()
```

```python
#Separate into data and target data
x = iris.data
y = iris.target
```

```python
#Converting numpy array into Pandas DataFrame
dfx = pd.DataFrame(data = x)
dfy = pd.DataFrame(data = y)
```

```python
#Create a singal data frame
dfz = dfx.join(dfy, lsuffix = "_dfx", rsuffix = "_dfy")
```

```python
#rename column headings
dfz.columns = [0,1,2,3,4,]
```

```python
#Hasn't been trained yet so column 4 will be left empty
dfz.head(5)
```

```python
import seaborn as sns
iris = sns.load_dataset("iris")
sns.pairplot(iris, hue = "species", markers = ["o", "s", "d"])
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 1)
```

```python
print(knn)
```

```python
knn.fit(x,y)
```

```python
x_new = [[3,4,4,2], [4,4,3,2]]
```

```python
knn.predict(x_new)
```

```python
#We predict that the first is virgincia and the second is versicolor
```

```python
knn1 = KNeighborsClassifier(n_neighbors = 5)
```

```python
knn1.fit(x,y)
```

```python
knn1.predict(x_new)
```

```python
#We predict that both flowers are of the versicolor species
```
